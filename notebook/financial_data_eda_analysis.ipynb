```python
import sys
import os
import pandas as pd

# Get the absolute path to the 'scripts' directory
scripts_dir = os.path.abspath('../scripts')
if scripts_dir not in sys.path:
    sys.path.append(scripts_dir)

from stock_analysis import load_stock_data
from stock_analysis import apply_technical_indicators
from stock_analysis import calculate_sharpe_ratio
from stock_analysis import plot_stock_indicators
```

##### defining stock names and Data paths


## Load The Stock Data# Step 4: Process Each Stock Dataset and Calculate Metrics


```python
# Load data
BASE_PATH = "../data/yfinance_data/"
STOCK_FILES = {
    "AAPL": "AAPL_historical_data.csv",
    "AMZN": "AMZN_historical_data.csv",
    "GOOG": "GOOG_historical_data.csv",
    "META": "META_historical_data.csv",
    "MSFT": "MSFT_historical_data.csv",
    "NVDA": "NVDA_historical_data.csv",
    "TSLA": "TSLA_historical_data.csv"
}

dataframes, combined_df = load_stock_data(BASE_PATH, STOCK_FILES)
```

```python
sharpe_ratios = {}
for symbol, df in dataframes.items():
    df = apply_technical_indicators(df)
    sharpe_ratios[symbol] = calculate_sharpe_ratio(df)
    plot_stock_indicators(df, symbol)
```

```python
# Display Sharpe Ratios
print("Sharpe Ratios:")
for stock, ratio in sharpe_ratios.items():
    print(f"{stock}: {ratio:.2f}")
```

### **Interpretation of Results**

Here are the Sharpe Ratios for the given stocks:

| **Stock** | **Sharpe Ratio** |
|-----------|------------------|
| AAPL      | 0.61             |
| AMZN      | 0.77             |
| GOOG      | 0.85             |
| META      | 0.82             |
| MSFT      | 0.82             |
| NVDA      | 0.81             |
| TSLA      | 0.90             |

---

### **Higher Sharpe Ratio**
A higher Sharpe Ratio indicates **better risk-adjusted returns**, meaning the stock generates higher returns for each unit of risk taken. In this case:

- **TSLA** has the highest Sharpe Ratio (0.90), making it the **best-performing stock** on a risk-adjusted basis.
- **GOOG** and **META** also perform well, with Sharpe Ratios of **0.85** and **0.82**, respectively.

---

### **Lower Sharpe Ratio**
A lower Sharpe Ratio indicates that the return may not adequately compensate for the risk involved:

- **AAPL** has the lowest Sharpe Ratio (**0.61**), suggesting its risk-adjusted return is comparatively **less favorable**.

---

### **Key Insight**
While all stocks show **positive Sharpe Ratios** (above 0), the following insights can be drawn:

- **TSLA** offers the **best balance of return relative to risk**, making it ideal for risk-adjusted investors.
- **AAPL** is the least favorable among the group, with the lowest Sharpe Ratio.

Investors seeking risk-adjusted returns may consider prioritizing stocks with **higher Sharpe Ratios** such as **TSLA**, **GOOG**, and **META**.
